```python
import json

form_components = []
form_component_multi_inputs = []
def load_json_text(json_text):
    form_components.clear()
    columns_dict = json.loads(json_text)['data']['templateBody']
    for k, column in columns_dict.items():
        if k.startswith('column'):
            form_components.extend(column)
    form_component_multi_inputs.clear()
    for form_component in form_components:
        if 'type' in form_component and form_component['type'] == 'multi-input':
            form_component_multi_inputs.append(form_component)


load_json_text(json1)
form_component_multi_inputs
```
